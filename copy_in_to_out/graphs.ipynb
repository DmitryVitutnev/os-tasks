```python
x = np.array([64, 128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536])
y = np.array([16.586, 8.692, 4.683, 2.573, 1.585, 1.118, 0.828, 0.749, 0.707, 0.667, 0.676])
plt.figure(figsize=(20,10))
plt.plot(x,y)
plt.xlabel('buffer size')
plt.ylabel('system time')
plt.show()
```
